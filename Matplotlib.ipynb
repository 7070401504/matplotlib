# 3D plot 

```python
x=np.random.rand(20)
y=np.random.rand(20)
z=np.random.rand(20)
fig=plt.figure()
ax=fig.add_subplot(projection="3d")
ax.scatter(x,y,z, c="#FF5733",alpha=.9)
plt.xlabel("x axis")
plt.xlabel("y axis")
plt.show()
```

```python
sns.scatterplot(x=tips.total_bill, y=tips.tip,data=tips,hue="smoker")
```

```python
sns.scatterplot(x=tips.total_bill, y=tips.tip,data=tips,hue='size')
```

```python
sns.scatterplot(x=tips.total_bill, y=tips.tip,data=tips,style='size')
```
